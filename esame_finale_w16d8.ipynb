â€¢Si chiede poi per ogni continente di trovare il numero di casi totali avvenuti in quello stesso continente; si chiede di non considerare eventuali locazioni che nel dataset non appartengono ad alcun continente;

```python
# DataFrame contenente i continenti
covid_continente = covid.dropna(subset=['continent'])
covid_continente
```

```python
casi_per_continente = covid_continente.groupby("continent")["new_cases"].sum()
casi_per_continente_df = casi_per_continente.to_frame().reset_index()
casi_per_continente_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data=casi_per_continente_df, x="new_cases", y="continent", palette="magma")
plt.title("Totale casi COVID-19 per continente")
plt.xlabel("Totale casi")
plt.ylabel("Continente")
plt.show()
```

â€¢Selezionare i dati relativi all'Italia nel 2022, e mostrare con un grafico adeguato l'evoluzione del casi totali rispetto alla data. Mostrare poi con un grafico adeguato il numero di nuovi casi rispetto alla data (filtrare i dati se necessario). Mostrare infine un grafico che mostra l'andamento della somma cumulativa nuovi casi del 2022, e commentare se ci sono similaritÃ  con l'andamento del numero dei casi totali;

```python
covid["date"]= pd.to_datetime(covid["date"])
```

```python
italia=covid[(covid.location=="Italy")&(covid["date"].dt.year==2022)]
italia.head(20)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=italia, x="date", y="total_cases")
```

```python
italia.new_cases.replace(0, np.nan, inplace=True)
italia = italia.dropna(subset=["date"])
plt.figure(figsize=(20,10))
sns.lineplot(data=italia, x="date", y="new_cases")
```

```python
plt.figure(figsize=(20, 6))
sns.lineplot(data=italia, x="date", y="total_cases", label='Casi Totali')
italia_cum = italia['new_cases'].cumsum()
sns.lineplot(data=italia, x='date', y=italia_cum, label='Somma Cumulativa Nuovi Casi')
plt.xlabel('Data')
plt.ylabel('Numero di Casi')
plt.title('Confronto dei Casi Totali e Somma Cumulativa')
plt.legend()
plt.show()
```

In questo grafico vediamo come all'aumentare dei nuovi casi aumenta il totale casi(che praticamente il totale dei casi non e altro che la somma cumulativa dei nuovi casi). Osservando anche il DataFrame vediamo la stessa cosa, praticamente il totale dei casi e una somma cumulativa che aumenta al momento del inserimento dei dati per i nuovi casi(ogni 7 gg), del numero dei nuovi casi.

â€¢Riguardo le nazioni di Italia, Germania e Francia, mostrare in un boxplot la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU) da maggio 2022 (incluso) ad aprile 2023 (incluso), e scrivere un breve commento a riguardo;

```python
IGF = covid[covid['location'].isin(['Italy', 'Germany', 'France'])]
IGF = IGF[(IGF['date'] >= '2022-05-01') & (IGF['date'] <= '2023-04-30')]
IGF
```

```python
plt.figure(figsize = (10, 6))  
sns.boxplot(data = IGF, x = 'location', y = 'icu_patients')  
plt.title('Boxplot dei dati per Italia, Germania e Francia')
plt.xlabel('Paese')
plt.ylabel('Terapia Intensiva') 
plt.show()
```

```python
plt.figure(figsize = (10, 6))  
sns.violinplot(data = IGF, x = 'location', y = 'icu_patients')  
plt.title('Boxplot dei dati per Italia, Germania e Francia')
plt.xlabel('Paese')
plt.ylabel('Terapia Intensiva') 
plt.show()
```

Il boxplot mostra la distribuzione del numero di pazienti in terapia intensiva per Italia, Germania e Francia durante il periodo maggio 2022 aprile 2023. Possiamo vedere come Italia ha una distribuzione piu compatta rispetto alla Germania e alla Francia.

â€¢Riguardo le nazioni di Italia, Germania, Francia e Spagna nel 2023, mostrare la somma dei pazienti ospitalizzati per ognuna â€” se ci sono dati nulli, suggerire se puÃ² essere possibile gestirli tramite sostituzione o meno

```python
IGFS2023 = covid[(covid['location'].isin(['Italy', 'Germany', 'France', 'Spain'])) & (covid["date"].dt.year==2023)]
ospIGFS2023 = IGFS2023.groupby('location')['hosp_patients'].sum()
osp=ospIGFS2023.to_frame().reset_index()
osp
```
